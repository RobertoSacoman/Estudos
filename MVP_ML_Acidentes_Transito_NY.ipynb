# **Introdução**

## Acidentes de Trânsito em Nova Iorque: Um Panorama

Nova Iorque, uma das cidades mais populosas e movimentadas dos Estados Unidos, enfrenta desafios significativos relacionados à segurança no trânsito. Com milhões de residentes, trabalhadores e turistas compartilhando ruas e rodovias, o tráfego intenso e o comportamento dos motoristas, ciclistas e pedestres tornam a segurança viária uma prioridade crítica.

**Fatores Contribuintes para Acidentes**  
Os acidentes de trânsito na cidade estão associados a uma combinação de fatores:  
1. **Densidade de tráfego**: As ruas de Nova Iorque frequentemente ficam congestionadas, aumentando o risco de colisões, especialmente em cruzamentos movimentados.  
2. **Comportamento humano**: Dirigir distraído, excesso de velocidade e a desobediência às regras de trânsito são causas comuns de acidentes. Além disso, o uso do celular enquanto se dirige é um problema crescente.  
3. **Clima e condições das vias**: Durante o inverno, o gelo e a neve dificultam o controle dos veículos, contribuindo para colisões e derrapagens.  
4. **Conflito entre modais**: Nova Iorque tem um número significativo de pedestres e ciclistas, que frequentemente dividem o espaço com veículos motorizados, criando zonas de conflito em vias compartilhadas.  

Os acidentes de trânsito em Nova Iorque geram uma série de problemas que afetam não apenas os envolvidos diretamente, mas também a cidade como um todo. Esses impactos podem ser categorizados em várias áreas:

#### 1. **Perdas Humanas e Lesões**
O impacto mais grave é a perda de vidas e os ferimentos causados. Pedestres, ciclistas, motoristas e passageiros podem sofrer consequências que variam de ferimentos leves a traumas graves, muitas vezes resultando em internações prolongadas e reabilitação. Por exemplo, em cruzamentos movimentados de Manhattan, é comum que pedestres sejam vítimas devido à combinação de tráfego intenso e distração.

#### 2. **Impactos Econômicos**
Os acidentes geram altos custos financeiros, como:
- **Custos médicos**: Para tratar feridos, incluindo emergências e cirurgias.
- **Danos materiais**: Reparos em veículos, bicicletas e infraestrutura urbana.
- **Produtividade**: A perda de dias de trabalho para os envolvidos e atrasos no transporte público.

A título de comparação, um único acidente grave em uma avenida principal pode atrasar milhares de trabalhadores que dependem de ônibus ou táxis para se locomover.

#### 3. **Congestionamentos no Trânsito**
Mesmo acidentes menores podem causar longos engarrafamentos, especialmente em horários de pico. Nova Iorque já enfrenta uma circulação densa, e qualquer interrupção pode agravar o problema, atrasando entregas, transporte público e movimentação de emergência.

#### 4. **Sobrecarregamento dos Serviços Públicos**
Acidentes demandam a intervenção de equipes de emergência, como polícia, bombeiros e paramédicos. Isso pode desviar recursos de outras áreas da cidade e sobrecarregar hospitais, que já lidam com altas taxas de atendimento.

#### 5. **Consequências Emocionais e Sociais**
Além das perdas físicas, há impactos emocionais. Vítimas e suas famílias frequentemente enfrentam traumas psicológicos, como transtorno de estresse pós-traumático (TEPT). Comunidades locais também podem sentir os efeitos, especialmente em bairros onde acidentes são recorrentes, prejudicando a sensação de segurança.

#### 6. **Degradação da Infraestrutura Urbana**
Acidentes frequentemente danificam postes, semáforos, calçadas e outros elementos da infraestrutura viária. Reparos demandam tempo e dinheiro, além de atrapalharem o fluxo normal de pessoas e veículos.

```python
# Carregar o dataset
url = "https://raw.githubusercontent.com/RobertoSacoman/Estudos/main/NYC_Collisions_2023.csv"
dataset = pd.read_csv(url, sep=";")

# Mostrar as primeiras linhas e informações do dataset
dataset.head()
```

Com os dados analisados acima, vamos iniciar a preparação dos dados inicias:




```python
# Preparação do dataset para modelo de classificação

# Remover colunas irrelevantes
irrelevant_columns = ['Collision ID', 'Date', 'Time', 'Cross Street', 'Latitude', 'Longitude']
dataset = dataset.drop(columns=irrelevant_columns, errors='ignore')

# Tratar valores ausentes
# Substituir valores nulos em 'Borough' por 'Unknown' e nas outras colunas por 0 (onde faz sentido)
dataset['Borough'].fillna('Unknown', inplace=True)
numerical_columns = ['Persons Injured', 'Persons Killed', 'Pedestrians Injured', 'Pedestrians Killed',
                     'Cyclists Injured', 'Cyclists Killed', 'Motorists Injured', 'Motorists Killed']
dataset[numerical_columns] = dataset[numerical_columns].fillna(0)

# Criar uma nova coluna de alvo (target) para classificação
# Classificar os acidentes com base em gravidade
dataset['Severity'] = dataset.apply(
    lambda row: 'Grave' if (
        row['Persons Injured'] > 0 or row['Persons Killed'] > 0 or
        row['Pedestrians Injured'] > 0 or row['Pedestrians Killed'] > 0 or
        row['Cyclists Injured'] > 0 or row['Cyclists Killed'] > 0 or
        row['Motorists Injured'] > 0 or row['Motorists Killed'] > 0
    ) else 'Não Grave', axis=1
)

# Remover as colunas que foram utilizadas para criar o target
dataset = dataset.drop(columns=numerical_columns, errors='ignore')

# Codificar variáveis categóricas usando one-hot encoding
dataset_encoded = pd.get_dummies(dataset, columns=['Borough', 'Street Name', 'Contributing Factor', 'Vehicle Type'], drop_first=True)

# Separar features e target
X = dataset_encoded.drop(columns=['Severity'], errors='ignore')
y = dataset_encoded['Severity']

# Criar um resumo das informações preparadas
data_summary = {
    "Número de linhas": dataset_encoded.shape[0],
    "Número de colunas": dataset_encoded.shape[1],
    "Colunas de features (amostra)": X.columns.tolist()[:10],  # Exibir algumas features como exemplo
    "Classes de target": y.unique().tolist()
}

# Exibir os dados preparados para visualização
data_summary
```

## Refinamento dos dados:

Tendo em vista que o grande problema de NY é a quantidade de acidentes, depois de realizar a preparação inicial dos dados, iremos iniciar o refino dos dados para começar a preparar o nosso modelo.

```python
# Agrupar o número total de acidentes por rua (usando as colunas de 'Street Name_*')
street_columns = [col for col in dataset_encoded.columns if col.startswith('Street Name_')]
street_accident_counts = dataset_encoded[street_columns].sum(axis=0)

# Criar categorias de classificação para as ruas
bins = [0, street_accident_counts.quantile(0.33), street_accident_counts.quantile(0.66), street_accident_counts.max()]
labels = ['Baixa', 'Média', 'Alta']
street_accident_categories = pd.cut(street_accident_counts, bins=bins, labels=labels, include_lowest=True)

# Combinar as categorias com os nomes das ruas
street_analysis = pd.DataFrame({
    'Street Name': street_accident_counts.index.str.replace('Street Name_', ''),
    'Accident Count': street_accident_counts.values,
    'Category': street_accident_categories
})

# Ordenar por número de acidentes para visualização
street_analysis = street_analysis.sort_values(by='Accident Count', ascending=False)

# Exibir o resumo da análise de acidentes por rua
print("Análise de Quantidade de Acidentes por Rua (Top 5):")
print(street_analysis.head())

# Exibir categorias de classificação
print("\nCategorias de classificação:")
print(f"Baixa: Até {bins[1]:.0f} acidentes")
print(f"Média: Entre {bins[1]:.0f} e {bins[2]:.0f} acidentes")
print(f"Alta: Acima de {bins[2]:.0f} acidentes")

```

## Primeiro vamos fazer o split dos dados

## A divisão foi realizada considerando apenas as ruas mais afetada(90%) que é o foco do nosso objetivo.

### Aqui estão os detalhes:

- Tamanho do conjunto de treinamento: 9.541 amostras.
- Tamanho do conjunto de teste: 4.089 amostras.
- Proporção de classes no conjunto de treinamento:
    - Não Grave: 56,65%
    - Grave: 43,35%
- Proporção de classes no conjunto de teste:
    - Não Grave: 56,66%
    - Grave: 43,34%

- Total de ruas filtradas: 321 ruas com maior frequência de acidentes.

```python
# Identificando as ruas mais afetadas
# Calculando o número total de acidentes por rua
street_accident_counts = dataset['Street Name'].value_counts()

# Definindo um limiar para determinar "ruas mais afetadas"
threshold = street_accident_counts.quantile(0.90)  # Mantendo as ruas acima do 90%

# Filtrando somente as ruas mais afetadas
high_accident_streets = street_accident_counts[street_accident_counts >= threshold].index
filtered_dataset = dataset[dataset['Street Name'].isin(high_accident_streets)]

# Aaplicando o one-hot encoding com as ruas filtradas
filtered_dataset_encoded = pd.get_dummies(
    filtered_dataset,
    columns=['Borough', 'Street Name', 'Contributing Factor', 'Vehicle Type']
)

# Separando as features e o target
X_filtered = filtered_dataset_encoded.drop(columns=['Severity'])
y_filtered = filtered_dataset_encoded['Severity']

# Realizando a divisão com 70% para treinamento e 30% para teste
X_train_filtered, X_test_filtered, y_train_filtered, y_test_filtered = train_test_split(
    X_filtered, y_filtered, test_size=0.3, random_state=42, stratify=y_filtered
)

# Resumo da nova divisão
filtered_split_summary = {
    "Tamanho do conjunto de treinamento": X_train_filtered.shape[0],
    "Tamanho do conjunto de teste": X_test_filtered.shape[0],
    "Proporção de classes no treinamento": y_train_filtered.value_counts(normalize=True).to_dict(),
    "Proporção de classes no teste": y_test_filtered.value_counts(normalize=True).to_dict(),
    "Total de ruas filtradas": len(high_accident_streets)
}

filtered_split_summary
```

Agora, vamos aplicar uma técnica de escala nos nossos dados numéricos.

```python
# Inicializando o escalonador
scaler = StandardScaler()

# Aplicando o escalonamento apenas nas features do conjunto filtrado
X_train_scaled = scaler.fit_transform(X_train_filtered)
X_test_scaled = scaler.transform(X_test_filtered)

# Verificando o resultado do escalonamento
scaled_summary = {
    "Média (após escalonamento)": X_train_scaled.mean(axis=0).tolist()[:5],  # Amostra de médias das features
    "Desvio padrão (após escalonamento)": X_train_scaled.std(axis=0).tolist()[:5]  # Amostra de desvios padrão das features
}

scaled_summary
```

# Para treinar e comparar os algoritmos, usaremos 5 dos mais populares:

### Logistic Regression
### Random Forest Classifier
### Support Vector Machine (SVM)
### K-Nearest Neighbors (KNN)
### Decision Tree

Sem validação Cruzada

```python
# Lista dos modelos para comparação
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Support Vector Machine": SVC(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
}

# Dicionário para armazenar os resultados
results = {}

# Comparação de modelos
for name, model in models.items():
    # Treinar o modelo
    model.fit(X_train_scaled, y_train_filtered)

    # Fazer previsões
    y_pred = model.predict(X_test_scaled)

    # Avaliar o modelo
    accuracy = accuracy_score(y_test_filtered, y_pred)
    report = classification_report(y_test_filtered, y_pred, output_dict=True)

    # Salvar os resultados
    results[name] = {
        "Accuracy": accuracy,
        "Classification Report": report
    }

# Exibir resultados dos 5 modelos
print("Resultados Comparativos dos Modelos:")
print(pd.DataFrame({k: v["Accuracy"] for k, v in results.items()}, index=["Accuracy"]).T)

# Identificar o melhor modelo com base na acurácia
best_model_name = max(results, key=lambda x: results[x]['Accuracy'])
best_model_results = results[best_model_name]

# Exibir resultados dos 5 modelos
best_model_name, best_model_results
```

Com validação cruzada

```python
# Lista dos modelos para comparação
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Support Vector Machine": SVC(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
}

# Dicionário para armazenar os resultados
results = {}

# Comparação de modelos com validação cruzada
for name, model in models.items():
    # Aplicar validação cruzada com 5 folds
    cv_scores = cross_val_score(model, X_train_scaled, y_train_filtered, cv=5, scoring='accuracy')
    mean_accuracy = cv_scores.mean()

    # Salvar os resultados
    results[name] = {
        "Mean Accuracy": mean_accuracy,
        "CV Scores": cv_scores
    }

# Exibir resultados dos modelos
print("Resultados Comparativos dos Modelos com Validação Cruzada:")
results_df = pd.DataFrame({k: v["Mean Accuracy"] for k, v in results.items()}, index=["Mean Accuracy"]).T
print(results_df)

# Identificar o melhor modelo com base na média de acurácia
best_model_name = max(results, key=lambda x: results[x]['Mean Accuracy'])
best_model_results = results[best_model_name]

print("\nMelhor Modelo:")
print(f"Modelo: {best_model_name}")
print(f"Mean Accuracy: {best_model_results['Mean Accuracy']:.4f}")

best_model_name, best_model_results
```

## Agora vamos utilizar os hiperparâmetros na tentativa de melhorar os resultados

```python
# Definir os hiperparâmetros para ajuste
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularização
    'solver': ['liblinear', 'lbfgs'],  # Algoritmos de otimização
    'penalty': ['l2'],  # Penalidade regularizadora
    'max_iter': [100, 500, 1000]  # Número de iterações
}

# Inicializar o modelo Logistic Regression
logistic_model = LogisticRegression(random_state=42)

# Configurar o GridSearchCV
grid_search = GridSearchCV(
    estimator=logistic_model,
    param_grid=param_grid,
    scoring='accuracy',
    cv=3,
    verbose=1,
    n_jobs=-1  # Usar múltiplos núcleos para acelerar
)

# Executar o ajuste de hiperparâmetros
grid_search.fit(X_train_scaled, y_train_filtered)

# Resultados do melhor modelo
best_params = grid_search.best_params_
best_score = grid_search.best_score_

best_params, best_score
```
